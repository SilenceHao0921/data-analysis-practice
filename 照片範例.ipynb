```python
from PIL import Image
import numpy as np

arr=Image.open('face10.jpg')

array = np.array(arr, dtype=np.uint8)

arr = np.empty(array.shape,dtype=np.uint8)
arr[:,:,0] = array[ : , : , 0 ]
arr[:,:,1] = array[ : , : , 0 ]
arr[:,:,2] = array[ : , : , 0 ]


a1 = np.concatenate((arr,array),axis=1)

import matplotlib.pyplot as plt

plt.imshow(a1,cmap='gray', vmin=0, vmax=255)
plt.title("Color Image")
plt.show()
```

```python
# 製作四張色階不同並排的照片
# 讀圖並確保 shape 為 (高,寬,3) 
array = np.array(Image.open('face10.jpg'), dtype=np.uint8)

# 取 R、G、B channel，各自是灰階
r = array[:, :, 0]
g = array[:, :, 1]
b = array[:, :, 2]

# 全部resize到一樣大小（選擇較小的高寬，避免顯示裁切問題）
h, w = r.shape
min_size = min(h, w)
r = r[:min_size, :min_size]
g = g[:min_size, :min_size]
b = b[:min_size, :min_size]
color = array[:min_size, :min_size, :]

# 佈局
plt.figure(figsize=(12, 4))
for i, img in enumerate([r, g, b, color]):
    plt.subplot(1, 4, i+1)
    if i < 3:
        plt.imshow(img, cmap='gray', vmin=0, vmax=255)
        plt.title(['R', 'G', 'B'][i])
    else:
        plt.imshow(img)
        plt.title('Color')
    plt.axis('off')

plt.tight_layout()
plt.show()
```

```python
arr = np.empty((300,500), dtype=np.uint8)
arr[ : ,    :100] = 242
arr[ : , 100:200] = 217
arr[ : , 200:300] = 191
arr[ : , 300:400] = 166
arr[ : , 400:   ] = 127
# 顯示灰階圖片
plt.imshow(arr, cmap='gray', vmin=0, vmax=255)
plt.title("Grayscale Image")
plt.show()

hex_color = ["#FED172","#F3742B","#BB3A14","#612E37","#231650"]

rgb_color = []
for i in hex_color:
    r = int(i[1:3],16)
    g = int(i[3:5],16)
    b = int(i[5:7],16)
    rgb_color.append((r,g,b))
print(rgb_color)
```

```python
arr = np.empty((300,500,3), dtype=np.uint8)
arr[ : ,    :100 , : ] = rgb_color[0]
arr[ : , 100:200 , : ] = rgb_color[1]
arr[ : , 200:300 , : ] = rgb_color[2]
arr[ : , 300:400 , : ] = rgb_color[3]
arr[ : , 400:    , : ] = rgb_color[4]

import matplotlib.pyplot as plt
# 顯示灰階圖片
plt.imshow(arr, vmin=0, vmax=255)
plt.title("Color Image")
plt.show()
```
